```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

data = pd.read_csv("datos.txt", names = ["Voltaje", "Corriente"], sep = " ", comment = '#', header = None )

data
```

```python
fig, ax = plt.subplots ( ncols = 1, nrows = 1, figsize = (5, 4), dpi = 130 )

ax.plot ( 
    data[['Corriente']], data[['Voltaje']], 
    linewidth = 2, label = "Resistencia")

ax.set_ylabel("Voltaje [V]")
ax.set_xlabel("Corriente [A]")
ax.grid(color="gray", linestyle="--")

plt.legend ( bbox_to_anchor = (0.8, 1), loc = 'best' )

plt.show()

fig.savefig("grafico2.png", dpi=130)
```

```python
def ajusLineal(y):
	return 25.37*y+4.6

y=range(0, 11)

fig, ax = plt.subplots ( ncols = 1, nrows = 1,  dpi = 80 )

plt.plot(y, [ajusLineal(i) for i in y], color="red", label="Resistencia")
plt.grid(linestyle="--", color="gray")
plt.ylabel("Voltaje [V]", fontsize="12")
plt.xlabel("Corriente [A]", fontsize="12")
plt.title("Grfico de A v/s V con ajuste lineal", fontsize="12")

ax.set_ylim ( 0, 23 )
ax.set_xlim ( 0, 0.6 )

plt.legend ( bbox_to_anchor = (0.8, 1), loc = 'best' )

plt.show()

fig.savefig("grafico3.png", dpi=80)
```
